```python
def bowler_average(runs, wickets):
    if(wickets==0):
        return runs*2
    else:
        return runs/wickets
    
def bowler_sr(balls, wickets):
    if(wickets==0):
        return balls*2
    else:
        return balls/wickets
    
def bowler_er(balls, runs):
    if(runs==0):
        return 0
    else:
        return runs/balls * 6
    
def bowler_dot_p(balls, dots):
    if(dots==0):
        return 0
    else:
        return dots/balls*100
    
def bowler_boundary_p(balls, boundaries):
    if(boundaries==0):
        return 0
    else:
        return boundaries/balls*100
```

```python
def bowler_stats(df):
    
    balls = pd.DataFrame(df.groupby('bowler')['ball'].count())
    wides = pd.DataFrame(df.groupby('bowler')['wides'].sum())
    noballs = pd.DataFrame(df.groupby('bowler')['noballs'].sum())
    balls = balls.merge(wides, on='bowler').merge(noballs, on='bowler') 
    balls['ball'] = balls['ball'] - balls['wides'] - balls['noballs']
    balls.drop(['wides', 'noballs'], axis = 1, inplace=True)
    balls.columns = ['Legitimate balls']
    
    runs =  pd.DataFrame(df.groupby('bowler')['bowler_runs'].sum())
    wickets = pd.DataFrame(df.groupby('bowler')['is_bowlers_wicket'].sum())
    dots = pd.DataFrame(df.groupby('bowler')['is_dot_bowler'].sum())
    fours = pd.DataFrame(df.groupby('bowler')['is_four'].sum())
    sixes = pd.DataFrame(df.groupby('bowler')['is_six'].sum())
    boundaries = pd.DataFrame(df.groupby('bowler')['is_boundary'].sum())

    bowler = balls.merge(wides, on='bowler').merge(noballs, on='bowler').merge(runs, on='bowler').merge(wickets, on='bowler').merge(dots, on='bowler').merge(fours, on='bowler').merge(sixes, on='bowler').merge(boundaries, on='bowler')
    bowler.columns = ['Legitimate balls', 'Wides', 'No Balls', 'Runs', 'Wickets', 'Dots', 'Fours', 'Sixes', 'Boundaries']
    
    bowler['Average'] = bowler.apply(lambda x: (bowler_average(x['Runs'], x['Wickets'])), axis=1)
    bowler['Strike Rate'] = bowler.apply(lambda x: (bowler_sr(x['Legitimate balls'], x['Wickets'])), axis=1)
    bowler['Economy Rate'] = bowler.apply(lambda x: (bowler_er(x['Legitimate balls'], x['Runs'])), axis=1)
    bowler['Dot%'] = bowler.apply(lambda x: (bowler_dot_p(x['Legitimate balls'], x['Dots'])), axis=1)
    bowler['Boundary%'] = bowler.apply(lambda x: (bowler_boundary_p(x['Legitimate balls'], x['Boundaries'])), axis=1)

    bowler = bowler.round(decimals=2)
    return bowler
```

```python
delhi_powerplay = bowler_stats(delhi_pp)
delhi_middle_overs = bowler_stats(delhi_mo)
delhi_death_overs = bowler_stats(delhi_do)

ahmedabad_powerplay = bowler_stats(ahmedabad_pp)
ahmedabad_middle_overs = bowler_stats(ahmedabad_mo)
ahmedabad_death_overs = bowler_stats(ahmedabad_do)
```

```python
def topsis(df, avg_wt, sr_wt, er_wt, dot_wt, boundary_wt):
    df['calc_avg'] = df['Average'].apply(lambda x: x/np.sqrt((df['Average']**2).sum()))
    df['calc_sr'] = df['Strike Rate'].apply(lambda x: x/np.sqrt((df['Strike Rate']**2).sum()))
    df['calc_er'] = df['Economy Rate'].apply(lambda x: x/np.sqrt((df['Economy Rate']**2).sum()))
    df['calc_dot%'] = df['Dot%'].apply(lambda x: x/np.sqrt((df['Dot%']**2).sum()))
    df['calc_boundary%'] = df['Boundary%'].apply(lambda x: x/np.sqrt((df['Boundary%']**2).sum()))

    df['calc_avg'] = df['calc_avg'].apply(lambda x: avg_wt*x)
    df['calc_sr'] = df['calc_sr'].apply(lambda x: sr_wt*x)
    df['calc_er'] = df['calc_er'].apply(lambda x: er_wt*x)
    df['calc_dot%'] = df['calc_dot%'].apply(lambda x: dot_wt*x)
    df['calc_boundary%'] = df['calc_boundary%'].apply(lambda x: boundary_wt*x)

    pos_avg = df['calc_avg'].min()
    pos_sr = df['calc_sr'].min()
    pos_er = df['calc_er'].min()
    pos_dot = df['calc_dot%'].max()
    pos_boundary = df['calc_boundary%'].min()

    neg_avg = df['calc_avg'].max()
    neg_sr = df['calc_sr'].max()
    neg_er = df['calc_er'].max()
    neg_dot = df['calc_dot%'].min()
    neg_boundary = df['calc_boundary%'].max()


    df['positive_diff'] = df.apply(lambda x: np.sqrt(((x['calc_avg']-pos_avg)**2)+((x['calc_sr']-pos_sr)**2) + ((x['calc_er']-pos_er)**2) + ((x['calc_dot%']-pos_dot)**2) + ((x['calc_boundary%']-pos_boundary)**2)), axis=1)
    df['negative_diff'] = df.apply(lambda x: np.sqrt(((x['calc_avg']-neg_avg)**2)+((x['calc_sr']-neg_sr)**2) + ((x['calc_er']-neg_er)**2) + ((x['calc_dot%']-neg_dot)**2) + ((x['calc_boundary%']-neg_boundary)**2)), axis=1)

    df['score'] = df.apply(lambda x: x['negative_diff']/(x['positive_diff']+x['negative_diff']), axis=1)

    df.sort_values('score', ascending=False, inplace=True)
    
    return df
```

```python
delhi_powerplay = topsis(delhi_powerplay, pp_avg_wt, pp_sr_wt, pp_er_wt, pp_dot_wt, pp_boundary_wt)
delhi_middle_overs = topsis(delhi_middle_overs, mo_avg_wt, mo_sr_wt, mo_er_wt, mo_dot_wt, mo_boundary_wt)
delhi_death_overs = topsis(delhi_death_overs, do_avg_wt, do_sr_wt, do_er_wt, do_dot_wt, do_boundary_wt)

ahmedabad_powerplay = topsis(ahmedabad_powerplay, pp_avg_wt, pp_er_wt, pp_er_wt, pp_dot_wt, pp_boundary_wt)
ahmedabad_middle_overs = topsis(ahmedabad_middle_overs, mo_avg_wt, mo_sr_wt, mo_er_wt, mo_dot_wt, mo_boundary_wt)
ahmedabad_death_overs = topsis(ahmedabad_death_overs, do_avg_wt, do_sr_wt, do_er_wt, do_dot_wt, do_boundary_wt)
```

```python
delhi_combined['calc_pp'] = delhi_combined['score_x'].apply(lambda x: x/np.sqrt((delhi_combined['score_x']**2).sum()))
delhi_combined['calc_mo'] = delhi_combined['score_y'].apply(lambda x: x/np.sqrt((delhi_combined['score_y']**2).sum()))
delhi_combined['calc_do'] = delhi_combined['score'].apply(lambda x: x/np.sqrt((delhi_combined['score']**2).sum()))

delhi_combined['calc_pp'] = delhi_combined['calc_pp'].apply(lambda x: 0.33*x)
delhi_combined['calc_mo'] = delhi_combined['calc_mo'].apply(lambda x: 0.33*x)
delhi_combined['calc_do'] = delhi_combined['calc_do'].apply(lambda x: 0.33*x)

pos_pp = delhi_combined['calc_pp'].max()
pos_mo = delhi_combined['calc_mo'].max()
pos_do = delhi_combined['calc_do'].max()

neg_pp = delhi_combined['calc_pp'].min()
neg_mo = delhi_combined['calc_mo'].min()
neg_do = delhi_combined['calc_do'].min()

delhi_combined['positive_diff'] = delhi_combined.apply(lambda x: np.sqrt(((x['calc_pp']-pos_pp)**2)+((x['calc_mo']-pos_mo)**2) + ((x['calc_do']-pos_do)**2)), axis=1)
delhi_combined['negative_diff'] = delhi_combined.apply(lambda x: np.sqrt(((x['calc_pp']-neg_pp)**2)+((x['calc_mo']-neg_mo)**2) + ((x['calc_do']-neg_do)**2)), axis=1)

delhi_combined['total_score'] = delhi_combined.apply(lambda x: x['negative_diff']/(x['positive_diff']+x['negative_diff']), axis=1)

delhi_combined.sort_values('total_score', ascending=False, inplace=True)
delhi_combined
```
